1. What is hypothesis testing in statistics ?
  - Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data. It helps determine whether there is enough evidence to support a specific belief or claim (called a hypothesis) about a population parameter.
2. What is the null hypothesis, and how does it differ from the alternative hypothesis ?
  - The null hypothesis (H₀) assumes no change or effect—it is the hypothesis tested directly.

  - The alternative hypothesis (H₁) suggests a change or effect—it is supported only if the evidence is strong enough to reject H₀.
3. What is the significance level in hypothesis testing, and why is it important?
  - The significance level (α) is the probability of making a Type I error, i.e., rejecting the null hypothesis when it is actually true.
4. What does a P-value represent in hypothesis testing ?
   - The p-value is the probability of obtaining a result as extreme as, or more extreme than, the observed result, assuming that the null hypothesis is true.
5. How do you interpret the P-value in hypothesis testing?
  - The p-value helps you decide whether to reject or fail to reject the null hypothesis (H₀) by measuring the strength of evidence against it.
6. What are Type 1 and Type 2 errors in hypothesis testing?
  - In hypothesis testing, errors can occur when making a decision to either reject or fail to reject the null hypothesis (H₀). These errors are categorized into two types:
      - Type 1 Error (α Error) : A Type 1 error occurs when the null hypothesis (H₀) is rejected when it is actually true.
      - Type 2 Error (β Error) : A Type 2 error occurs when the null hypothesis (H₀) is not rejected when it is actually false.
7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  - In hypothesis testing, the choice between a one-tailed and a two-tailed test depends on the research question and the direction of the effect you are testing for.
     - One-Tailed Test : Tests whether a parameter is greater than or less than a certain value.
     - Two-Tailed Test : Tests whether a parameter is different from a certain value (could be either higher or lower)
8. What is the Z-test, and when is it used in hypothesis testing?
  - The Z-test is a statistical test used to determine whether there is a significant difference between sample and population means or between two sample means when the population variance is known and the sample size is large (typically n ≥ 30).

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
  - The Z-score is a standardized value that tells you how many standard deviations a data point (or sample mean) is from the population mean. It is used in the Z-test to assess statistical significance.
10. What is the T-distribution, and when should it be used instead of the normal distribution?
  - The T-distribution (or Student’s T-distribution) is a probability distribution used in hypothesis testing when the sample size is small and the population standard deviation (σ) is unknown.
11. What is the difference between a Z-test and a T-test?
  - Use a Z-test when the population variance is known and the sample size is large enough for the Central Limit Theorem to apply.
  - Use a T-test when the population variance is unknown and the sample size is small, relying on the sample variance instead.
12.  What is the T-test, and how is it used in hypothesis testing?
  - The T-test is a statistical method used in hypothesis testing to determine if there is a significant difference between the means of two groups when the population standard deviation is unknown and/or the sample size is small.
13. What is the relationship between Z-test and T-test in hypothesis testing?
  - The Z-test is a special case of the T-test used when population variance is known or sample size is large.

  - The T-test extends the Z-test to cases where population variance is unknown and sample sizes are small.

  - For large samples, both tests will yield nearly the same conclusions.
14. What is a confidence interval, and how is it used to interpret statistical results?
  - A confidence interval gives a range of values within which the true population parameter likely falls, based on sample data. It helps interpret statistical results by showing the estimate’s uncertainty and guiding decisions about hypotheses.
15. What is the margin of error, and how does it affect the confidence interval?
  - The margin of error is a statistic that expresses the amount of random sampling error in a survey's results or an estimate. It quantifies how much the estimated value (like a sample mean or proportion) is expected to vary from the true population value due to chance.
16. How is Bayes' Theorem used in statistics, and what is its significance?
  - Bayes’ Theorem is crucial for updating probabilities in light of new evidence, making it a cornerstone of modern statistics, especially in fields involving uncertainty and decision making.
17. What is the Chi-square distribution, and when is it used?
  - The Chi-square distribution is mainly used for tests involving categorical data (like goodness-of-fit and independence tests) and for inference on variance when data is normally distributed. It provides the theoretical foundation for many common statistical tests involving frequencies and variance estimation.
18. What is the Chi-square goodness of fit test, and how is it applied?
  - The Chi-square goodness of fit test is a straightforward way to check if your observed categorical data matches an expected distribution by comparing observed vs. expected counts, calculating the Chi-square statistic, and assessing significance.
19. What is the F-distribution, and when is it used in hypothesis testing?
  - The F-distribution models the ratio of two scaled variances (chi-square variables divided by their df).

  - It is widely used in testing hypotheses about variances and in ANOVA to compare means across multiple groups.

  - Its key role is in deciding whether observed differences in sample variances or means are statistically significant.
20. What is an ANOVA test, and what are its assumptions?
  - ANOVA (Analysis of Variance) is a statistical method used to test whether there are significant differences between the means of three or more independent groups. Instead of comparing means pairwise, ANOVA checks if at least one group mean is different from the others.
21. What are the different types of ANOVA tests?
  - One-Way ANOVA : Tests differences between the means of three or more independent groups based on one factor (independent variable).
  - Two-Way ANOVA (Factorial ANOVA) : Tests the effect of two independent factors on a dependent variable and can also test for an interaction effect between the two factors.
22. What is the F-test, and how does it relate to hypothesis testing?
  - The F-test evaluates whether observed variances or explained variances differ more than expected by chance.
 - It’s fundamental for testing differences among group means (ANOVA) and assessing overall model significance (regression).

  - The F-test uses the F-distribution and compares an observed F-statistic against a critical value to make decisions about hypotheses.

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
def bernoulli_pmf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
    x : int
        The outcome (0 or 1)
    p : float
        Probability of success (0 <= p <= 1)

    Returns:
    float
        Probability of observing x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.7
print(f"P(X=1) = {bernoulli_pmf(1, p)}")
print(f"P(X=0) = {bernoulli_pmf(0, p)}")
```

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for the discrete uniform distribution
a = 1
b = 6
values = np.arange(a, b + 1)

# Probability for each value (equal probability)
prob = 1 / len(values)

# Calculate PMF
pmf = np.full(len(values), prob)

# Calculate CDF by cumulative sum of PMF
cdf = np.cumsum(pmf)

# Plot CDF as a step plot
plt.step(values, cdf, where='post', label='CDF', linewidth=2)

# Plot formatting
plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.xticks(values)
plt.ylim(0, 1.05)
plt.grid(True)
plt.legend()
plt.show()
```

9. Write a Python function to calculate Z-scores from a dataset and plot them

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores for a given dataset and plot them.

    Parameters:
    data (array-like): Input dataset (list, numpy array, etc.)

    Returns:
    z_scores (numpy array): Calculated Z-scores
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.stem(range(len(data)), z_scores, basefmt=" ", use_line_collection=True)
    plt.title('Z-scores of the Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage
sample_data = [10, 15, 14, 18, 20, 22, 21, 19, 17, 16]
z_scores = calculate_and_plot_z_scores(sample_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Parameters
population_size = 100000
sample_sizes = [1, 5, 30, 50]  # Different sample sizes to demonstrate CLT
num_samples = 1000             # Number of samples for each sample size

# Generate population from a non-normal distribution (Exponential)
population = np.random.exponential(scale=1.0, size=population_size)

plt.figure(figsize=(14, 10))

# Plot the original population distribution
plt.subplot(2, 2, 1)
sns.histplot(population, bins=50, kde=True, color='skyblue')
plt.title('Population Distribution (Exponential)')
plt.xlabel('Value')

# For each sample size, sample means and plot
for i, n in enumerate(sample_sizes, start=2):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=n, replace=True)
        sample_means.append(np.mean(sample))
    sample_means = np.array(sample_means)

    plt.subplot(2, 2, i)
    sns.histplot(sample_means, bins=30, kde=True, stat='density', color='lightgreen')

    # Overlay normal distribution for comparison
    mu, sigma = np.mean(sample_means), np.std(sample_means)
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
    plt.plot(x, norm.pdf(x, mu, sigma), 'r--', label='Normal PDF')

    plt.title(f'Sample Means Distribution (n={n})')
    plt.xlabel('Sample Mean')
    plt.legend()

plt.tight_layout()
plt.show()
```

14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, data):
    """
    Calculate the Z-score of a data point relative to the dataset.

    Parameters:
    - data_point: the value to calculate the Z-score for
    - data: list or numpy array of dataset values

    Returns:
    - z_score: the Z-score of the data_point
    """
    mean = np.mean(data)
    std = np.std(data)
    z_score = (data_point - mean) / std
    return z_score

# Example dataset and data point
dataset = [10, 15, 14, 18, 20, 22, 21, 19, 17, 16]
point = 18

# Calculate Z-score
z = calculate_z_score(point, dataset)
print(f"Z-score of {point}: {z:.3f}")

# Plot the standard normal distribution
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, 0, 1)

plt.plot(x, pdf, label='Standard Normal Distribution')
plt.fill_between(x, pdf, alpha=0.2)

# Mark the Z-score on the plot
plt.axvline(z, color='red', linestyle='--', label=f'Z-score = {z:.2f}')
plt.title('Standard Normal Distribution with Z-score')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

15.  Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, mu_0, alpha=0.05):
    """
    Perform a two-tailed Z-test for the sample mean against population mean mu_0.

    Parameters:
    - sample: list or numpy array of sample data
    - mu_0: hypothesized population mean under H0
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: calculated Z-statistic
    - p_value: two-tailed p-value
    - conclusion: string stating whether to reject or fail to reject H0
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # sample std dev (unbiased)

    # For Z-test, assume population std is known or use sample std if large n
    # Here, we'll approximate with sample std / sqrt(n)
    se = sample_std / np.sqrt(n)

    # Calculate Z-statistic
    z_stat = (sample_mean - mu_0) / se

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Decision
    if p_value < alpha:
        conclusion = "Reject the null hypothesis."
    else:
        conclusion = "Fail to reject the null hypothesis."

    return z_stat, p_value, conclusion

# Example sample data
sample_data = [52, 48, 49, 51, 47, 50, 53, 46, 54, 49]

# Hypothesized population mean
mu_hypothesis = 50

# Perform Z-test
z_statistic, p_val, result = z_test(sample_data, mu_hypothesis)

print(f"Z-statistic: {z_statistic:.3f}")
print(f"P-value: {p_val:.4f}")
print(f"Conclusion: {result}")
```

16. Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of the dataset.

    Parameters:
    - data: list or numpy array of data points
    - confidence: confidence level (default 95%)

    Returns:
    - (lower_bound, upper_bound): tuple of confidence interval limits
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)  # t critical value

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example dataset
sample_data = [12, 15, 14, 16, 15, 14, 13, 16, 14, 15]

# Calculate 95% confidence interval
ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Interpretation
print(f"This means we are 95% confident that the true population mean lies between {ci_lower:.2f} and {ci_upper:.2f}.")
```

17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

```python
import numpy as np
from scipy import stats

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # For reproducibility
mu_true = 100       # True mean
sigma_true = 15     # True standard deviation
sample_size = 50

data = np.random.normal(mu_true, sigma_true, sample_size)

# Step 2: Calculate the confidence interval for the mean
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)  # t critical value

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return mean, lower_bound, upper_bound

mean, ci_lower, ci_upper = confidence_interval(data, confidence=0.95)

# Step 3: Print results and interpretation
print(f"Sample mean: {mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print(f"Interpretation: We are 95% confident that the true population mean lies between {ci_lower:.2f} and {ci_upper:.2f}.")
```

21. Write a Python program to compare the standard deviations of two datasets and visualize the difference

```python
import numpy as np
import matplotlib.pyplot as plt

# Example datasets
data1 = np.array([12, 15, 14, 16, 15, 14, 13, 16, 14, 15])
data2 = np.array([22, 25, 19, 24, 23, 20, 21, 26, 22, 24])

# Calculate standard deviations
std1 = np.std(data1, ddof=1)  # Sample std dev
std2 = np.std(data2, ddof=1)

# Calculate means for plotting
mean1 = np.mean(data1)
mean2 = np.mean(data2)

# Prepare bar chart data
labels = ['Dataset 1', 'Dataset 2']
means = [mean1, mean2]
stds = [std1, std2]

# Plot
fig, ax = plt.subplots()
bars = ax.bar(labels, means, yerr=stds, capsize=10, color=['skyblue', 'lightgreen'])

# Add title and labels
ax.set_ylabel('Mean Value')
ax.set_title('Comparison of Standard Deviations Between Two Datasets')

# Show standard deviation values on top of bars
for bar, std in zip(bars, stds):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height + 0.5, f'Std: {std:.2f}',
            ha='center', va='bottom')

plt.show()
```

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Perform a one-sample Z-test for the mean.

    Parameters:
    - sample_data: array-like, sample observations
    - pop_mean: float, known population mean
    - pop_std: float, known population standard deviation
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: calculated Z statistic
    - p_value: two-tailed p-value
    - conclusion: string interpretation of the test
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate standard error
    se = pop_std / np.sqrt(n)

    # Calculate Z statistic
    z_stat = (sample_mean - pop_mean) / se

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Interpret results
    if p_value < alpha:
        conclusion = f"Reject the null hypothesis at α={alpha}. Evidence suggests the sample mean differs from population mean."
    else:
        conclusion = f"Fail to reject the null hypothesis at α={alpha}. No sufficient evidence to say sample mean differs."

    return z_stat, p_value, conclusion

# Example usage
np.random.seed(42)
population_mean = 100
population_std = 15

# Generate a sample dataset
sample = np.random.normal(loc=102, scale=population_std, size=30)

z_stat, p_val, result = z_test(sample, population_mean, population_std)

print(f"Z statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.4f}")
print("Interpretation:", result)
```

2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python

```python
import numpy as np
from scipy.stats import norm

def two_sample_z_test(sample1, sample2, pop_std1, pop_std2, alpha=0.05):
    """
    Perform two-sample Z-test for means with known population std deviations.

    Parameters:
    - sample1, sample2: arrays, sample data
    - pop_std1, pop_std2: floats, known population std deviations
    - alpha: significance level

    Returns:
    - z_stat: Z test statistic
    - p_value: two-tailed p-value
    - conclusion: interpretation string
    """
    n1, n2 = len(sample1), len(sample2)
    mean1, mean2 = np.mean(sample1), np.mean(sample2)

    # Standard error of difference
    se_diff = np.sqrt(pop_std1**2 / n1 + pop_std2**2 / n2)

    # Z statistic
    z_stat = (mean1 - mean2) / se_diff

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    if p_value < alpha:
        conclusion = f"Reject null hypothesis: means differ (p={p_value:.4f})"
    else:
        conclusion = f"Fail to reject null hypothesis: no evidence means differ (p={p_value:.4f})"

    return z_stat, p_value, conclusion

# Simulate random data
np.random.seed(2025)
pop_std_1, pop_std_2 = 5, 7
sample_1 = np.random.normal(loc=50, scale=pop_std_1, size=40)
sample_2 = np.random.normal(loc=53, scale=pop_std_2, size=35)

# Perform test
z_stat, p_val, result = two_sample_z_test(sample_1, sample_2, pop_std_1, pop_std_2)

print(f"Z statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.4f}")
print(result)
```

4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    se = pop_std / np.sqrt(n)
    z_stat = (sample_mean - pop_mean) / se

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Critical values for rejection regions
    z_critical_low = norm.ppf(alpha / 2)
    z_critical_high = norm.ppf(1 - alpha / 2)

    # Plotting
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Shade rejection regions
    plt.fill_between(x, 0, y, where=(x <= z_critical_low), color='red', alpha=0.3, label='Rejection Region (left)')
    plt.fill_between(x, 0, y, where=(x >= z_critical_high), color='red', alpha=0.3, label='Rejection Region (right)')

    # Plot z_stat
    plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z statistic = {z_stat:.3f}')

    # Labels and legend
    plt.title('Two-tailed Z-test with Rejection Regions')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Interpretation
    if z_stat < z_critical_low or z_stat > z_critical_high:
        conclusion = f"Reject null hypothesis (z = {z_stat:.3f}, p = {p_value:.4f})"
    else:
        conclusion = f"Fail to reject null hypothesis (z = {z_stat:.3f}, p = {p_value:.4f})"

    return z_stat, p_value, conclusion

# Example usage
np.random.seed(10)
pop_mean = 50
pop_std = 8
sample_data = np.random.normal(loc=53, scale=8, size=30)

z_stat, p_val, conclusion = two_tailed_z_test(sample_data, pop_mean, pop_std)

print(f"Z statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.4f}")
print("Conclusion:", conclusion)
```

5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(pop_mean, pop_std, n, alpha, alt_mean):
    """
    Visualize Type I and Type II errors for a two-tailed Z-test.

    Parameters:
    - pop_mean: mean under null hypothesis (H0)
    - pop_std: population standard deviation (known)
    - n: sample size
    - alpha: significance level (Type I error rate)
    - alt_mean: mean under alternative hypothesis (H1)
    """
    se = pop_std / np.sqrt(n)

    # Critical z-values for two-tailed test
    z_crit_low = norm.ppf(alpha / 2)
    z_crit_high = norm.ppf(1 - alpha / 2)

    # Critical values in sample mean scale
    crit_low = pop_mean + z_crit_low * se
    crit_high = pop_mean + z_crit_high * se

    # Type II error (β) calculation:
    # Probability under alternative distribution that sample mean falls inside [crit_low, crit_high]
    beta = norm.cdf(crit_high, loc=alt_mean, scale=se) - norm.cdf(crit_low, loc=alt_mean, scale=se)

    power = 1 - beta

    # Prepare x range for plotting
    x_min = min(pop_mean - 4*se, alt_mean - 4*se)
    x_max = max(pop_mean + 4*se, alt_mean + 4*se)
    x = np.linspace(x_min, x_max, 1000)

    # Null and alternative distributions
    null_pdf = norm.pdf(x, loc=pop_mean, scale=se)
    alt_pdf = norm.pdf(x, loc=alt_mean, scale=se)

    plt.figure(figsize=(12, 6))

    # Plot null distribution
    plt.plot(x, null_pdf, label='Null Hypothesis (H0)', color='blue')
    # Shade rejection regions (Type I error)
    plt.fill_between(x, 0, null_pdf, where=(x <= crit_low), color='red', alpha=0.3, label='Type I Error Region (α)')
    plt.fill_between(x, 0, null_pdf, where=(x >= crit_high), color='red', alpha=0.3)

    # Plot alternative distribution
    plt.plot(x, alt_pdf, label='Alternative Hypothesis (H1)', color='green')
    # Shade Type II error region
    plt.fill_between(x, 0, alt_pdf, where=((x > crit_low) & (x < crit_high)), color='orange', alpha=0.3, label='Type II Error Region (β)')

    # Draw critical lines
    plt.axvline(crit_low, color='black', linestyle='--', label='Critical Values')
    plt.axvline(crit_high, color='black', linestyle='--')

    plt.title('Type I (α) and Type II (β) Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Significance level (Type I error, α): {alpha}")
    print(f"Type II error (β) for alternative mean {alt_mean}: {beta:.4f}")
    print(f"Power (1 - β): {power:.4f}")

# Example usage
plot_type1_type2_errors(pop_mean=100, pop_std=15, n=50, alpha=0.05, alt_mean=108)
```

6. Write a Python program to perform an independent T-test and interpret the results

```python
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(sample1, sample2, alpha=0.05):
    """
    Perform an independent two-sample t-test and interpret results.

    Parameters:
    - sample1, sample2: arrays or lists of sample data
    - alpha: significance level

    Returns:
    - t_stat: t statistic
    - p_value: p-value
    - conclusion: string interpreting the test
    """
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)  # Welch's t-test (does not assume equal variances)

    if p_value < alpha:
        conclusion = f"Reject the null hypothesis: samples have significantly different means (p = {p_value:.4f})"
    else:
        conclusion = f"Fail to reject the null hypothesis: no significant difference in means (p = {p_value:.4f})"

    return t_stat, p_value, conclusion

# Example usage
np.random.seed(42)
group1 = np.random.normal(loc=20, scale=5, size=30)
group2 = np.random.normal(loc=22, scale=5, size=30)

t_stat, p_val, interpretation = independent_t_test(group1, group2)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_val:.4f}")
print("Interpretation:", interpretation)
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean using the t-distribution.

    Parameters:
    - sample: array-like, sample data
    - confidence: confidence level (default 0.95 for 95% CI)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # sample standard error
    dof = n - 1  # degrees of freedom

    # t critical value for two-tailed CI
    t_crit = t.ppf((1 + confidence) / 2, dof)

    margin_of_error = t_crit * std_err

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=40)

ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)

print(f"Sample mean: {np.mean(sample_data):.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data

```python
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for the sample mean at a given confidence level.

    Parameters:
    - sample: array-like, sample data
    - confidence: confidence level (default 0.95)

    Returns:
    - margin_of_error: the margin of error value
    """
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # standard error of the mean
    dof = n - 1  # degrees of freedom

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, dof)

    margin = t_crit * std_err
    return margin

# Example usage
np.random.seed(123)
sample_data = np.random.normal(loc=75, scale=12, size=50)

moe = margin_of_error(sample_data, confidence=0.95)
sample_mean = np.mean(sample_data)

print(f"Sample mean: {sample_mean:.2f}")
print(f"Margin of Error at 95% confidence: ±{moe:.2f}")
print(f"Confidence Interval: ({sample_mean - moe:.2f}, {sample_mean + moe:.2f})")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayesian_inference(prior_disease, sensitivity, false_positive_rate):
    """
    Calculate the posterior probability P(Disease | Positive) using Bayes' Theorem.

    Parameters:
    - prior_disease: P(Disease), prior probability of disease
    - sensitivity: P(Positive | Disease), test sensitivity
    - false_positive_rate: P(Positive | No Disease), false positive rate

    Returns:
    - posterior: P(Disease | Positive), updated probability given positive test
    """
    prior_no_disease = 1 - prior_disease

    # Total probability of positive test
    prob_positive = sensitivity * prior_disease + false_positive_rate * prior_no_disease

    # Bayes' Theorem
    posterior = (sensitivity * prior_disease) / prob_positive

    return posterior

# Example values
prior_disease = 0.01          # 1% prior chance of disease
sensitivity = 0.99            # 99% chance test is positive if disease present
false_positive_rate = 0.05    # 5% chance test is positive if no disease

posterior_probability = bayesian_inference(prior_disease, sensitivity, false_positive_rate)
print(f"Posterior probability of having the disease given a positive test: {posterior_probability:.4f}")
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for Chi-square test.

    Parameters:
    - observed: 2D numpy array representing observed frequencies

    Returns:
    - expected: 2D numpy array of expected frequencies
    """
    # Row sums and column sums
    row_sums = observed.sum(axis=1, keepdims=True)
    col_sums = observed.sum(axis=0, keepdims=True)
    total = observed.sum()

    # Expected frequency = (row sum * column sum) / total
    expected = (row_sums @ col_sums) / total
    return expected

# Example observed data (contingency table)
observed = np.array([[30, 10, 20],
                     [20, 25, 15]])

expected = calculate_expected_frequencies(observed)

print("Observed Frequencies:")
print(observed)
print("\nExpected Frequencies:")
print(expected)
```
